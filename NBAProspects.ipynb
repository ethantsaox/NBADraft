```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

```python
url = 'https://basketball.realgm.com/nba/draft/prospects/stats'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')
```

```python
table = soup.find('table', class_ = 'tablesaw compact')
```

```python
print(table)
```

```python
world_titles = soup.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

```python
df = pd.DataFrame(columns = world_table_titles)

df
```

```python
df = df.drop(columns=['7:00 PM ET', 'Preview\n\xa0\nRegular Season', '7:30 PM ET', '8:00 PM ET'])

df
```

```python
soup.find_all('tr')
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[2:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv('103NBAProspects.csv', index = False)
```
