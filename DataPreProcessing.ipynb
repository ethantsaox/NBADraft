```python
import pandas as pd

# Load the first DataFrame
df = pd.read_csv("/Users/ethantsao/NBADraft1/nbaplayersdraft.csv")

# Rename columns to match the second DataFrame
df_processed = df.rename(columns={
    'player': 'Player',
    'team': 'Team',
    'games': 'GP',
    'average_minutes_played': 'MPG',
    'points_per_game': 'PPG',
    'field_goal_percentage': 'FG%',
    '3_point_percentage': '3P%',
    'free_throw_percentage': 'FT%',
    'average_total_rebounds': 'RPG',
    'average_assists': 'APG',
    # Add other columns if available in the first DataFrame
})

# Calculate missing columns if necessary
# For example, if 'FGM' (field goals made) and 'FGA' (field goals attempted) are not available,
# you can calculate them using 'points_per_game' and 'field_goal_percentage'.
# However, since these columns are not directly available in the first DataFrame, we'll skip them.

# Drop unnecessary columns
columns_to_keep = [
    'Player', 'Team', 'GP', 'MPG', 'PPG', 'FG%', '3P%', 'FT%', 'RPG', 'APG', 'overall_pick'
]
df_processed = df_processed[columns_to_keep]

# Handle missing values
df_processed = df_processed.fillna({
    'GP': df_processed['GP'].mean(),
    'MPG': df_processed['MPG'].mean(),
    'PPG': df_processed['PPG'].mean(),
    'FG%': df_processed['FG%'].mean(),
    '3P%': df_processed['3P%'].mean(),
    'FT%': df_processed['FT%'].mean(),
    'RPG': df_processed['RPG'].mean(),
    'APG': df_processed['APG'].mean()
})

df_processed = df_processed.round(3)
```
